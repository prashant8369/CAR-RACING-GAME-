```python
import pygame
import time
import random

pygame.init()

# Define some colors
BLACK = (0, 0, 0)
WHITE = (255, 255, 255)
GREEN = (0, 255, 0)
RED = (255, 0, 0)

# Set the width and height of the screen [width, height]
SCREEN_WIDTH = 800
SCREEN_HEIGHT = 600
screen = pygame.display.set_mode((SCREEN_WIDTH, SCREEN_HEIGHT))

pygame.display.set_caption("Car Racing Game")

# Create player object
player_width = 73
player_height = 82
player_img = pygame.image.load("car.png")
player_img = pygame.transform.scale(player_img, (player_width, player_height))

def player(x, y):
    screen.blit(player_img, (x, y))

# Create enemy object
enemy_width = 70
enemy_height = 120
enemy_img = pygame.image.load("enemy.png")
enemy_img = pygame.transform.scale(enemy_img, (enemy_width, enemy_height))

def enemy(x, y):
    screen.blit(enemy_img, (x, y))

# Set the game clock
clock = pygame.time.Clock()

def game_loop():
    # Set the player's starting position
    player_x = (SCREEN_WIDTH * 0.45)
    player_y = (SCREEN_HEIGHT * 0.8)

    # Set the enemy's starting position
    enemy_x = random.randrange(0, SCREEN_WIDTH - enemy_width)
    enemy_y = -600
    enemy_speed = 5

    # Set the player's movement speed
    player_speed = 0
    player_max_speed = 5
    player_min_speed = -5

    # Set the score
    score = 0

    # Set the game over flag
    game_over = False

    while not game_over:
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                game_over = True
                pygame.quit()
                quit()

            # Check for key presses
            if event.type == pygame.KEYDOWN:
                if event.key == pygame.K_LEFT:
                    player_speed = player_min_speed
                elif event.key == pygame.K_RIGHT:
                    player_speed = player_max_speed

            # Check for key releases
            if event.type == pygame.KEYUP:
                if event.key == pygame.K_LEFT or event.key == pygame.K_RIGHT:
                    player_speed = 0

        # Move the player
        player_x += player_speed

        # Move the enemy
        enemy_y += enemy_speed

        # Check for collision
        if player_y < enemy_y + enemy_height:
            if player_x > enemy_x and player_x < enemy_x + enemy_width or player_x + player_width > enemy_x and player_x + player_width < enemy_x + enemy_width:
                game_over = True

        # Check if enemy has passed the player
        if enemy_y > SCREEN_HEIGHT:
            enemy_y = 0 - enemy_height
            enemy_x = random.randrange(0, SCREEN_WIDTH - enemy_width)
            score += 1
            enemy_speed += 1

        # Draw the player, enemy, and score
        screen.fill(WHITE)
        player(player_x, player_y)
        enemy(enemy_x, enemy_y)
        font = pygame.font.Font(None, 36)
        text = font.render("Score: " + str(score), True, BLACK)
        screen.blit(text, (10, 10))

        # Update the screen
        pygame.display.update()

    # Set the frame rate
    clock.tick(60)

# Game over message
font = pygame.font.Font(None, 72)
text = font.render("Game Over", True, RED)
text_rect = text.get_rect()
text_rect.center = (SCREEN_WIDTH // 2, SCREEN_HEIGHT // 2)
screen.blit(text, text_rect)
pygame.display.update()

# Wait for 3 seconds before closing the game
time.sleep(3)

# Quit pygame and exit the program
pygame.quit()
quit()

```
